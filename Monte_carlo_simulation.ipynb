$$Monte Carlo Simulation$$
$$ x ∼ N(0,1)$$
$$ E(sin(x)) = ?$$
$$ E(sin(x)) ≈ \frac{∑_{i=1}^{N}\sin(x_i)}{N}$$

* N이 크면 클 수록 더 정확한 근사치를 얻을 수 있음

```python
import numpy as np
```

```python
N = 10
x_i = np.random.standard_normal()
print(x_i)
print(np.sin(x_i))
```

```python
N = 1000000
sum = 0
for i in range(N):
  x_i = np.random.standard_normal() # Simulate a random number from standard normal distribution
  sum += np.sin(x_i)

mc_output = sum / N # Compute average
print(mc_output) # print monte carlo simulation output
```

$$\epsilon \sim N(2,3^2)$$
$$ f(\epsilon) = \epsilon^2cos(\epsilon) + exp(\epsilon)$$
$$ E(f(ϵ)) = ?$$
$$ E(f(ϵ)) \approx \frac{\sum_{i=1}^N f(\epsilon_i)}{N}$$

```python
# Define function

def f(epsilon):
    return epsilon**2 * np.cos(epsilon) + np.exp(epsilon)

N = 100
mu = 2
sigma = 3

e = mu + sigma * np.random.standard_normal(N)
np.average(f(e))
```

# Monte Carlo Simulation for pi

```python
import numpy as np

N = 10000000 # the number of trials(simulation)
red = 0
blue = 0

for i in range(N):
  x = np.random.random()
  y = np.random.random()
  if x**2 + y**2 < 1:
    red += 1

print(red)
print('pi/4 is similar to', red/N)
print('pi is', 4 * red / N)
```

```python
N = 1000 # the number of trials(simulation)
inside = []
outside = []

for i in range(N):
  x = np.random.random()
  y = np.random.random()
  if x**2 + y**2 < 1:
    inside.append((x, y))
  else:
    outside.append((x, y))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (5, 5))
plt.scatter([x[0] for x in inside], [x[1] for x in inside], s = 5, c = 'red')
plt.scatter([x[0] for x in outside], [x[1] for x in outside], s = 5, c = 'blue')
plt.show()
```

```python
def generate_points(N):
    inside = []
    outside = []
    for i in range(N):
        x = np.random.random()
        y = np.random.random()
        if x**2 + y**2 < 1:
            inside.append((x, y))
        else:
            outside.append((x, y))
    return inside, outside

# 각 N 값에 대해 점을 생성
N_values = [100, 1000, 10000, 100000, 1000000]
results = [generate_points(N) for N in N_values]

# 그림 크기 및 서브플롯 설정
fig, axes = plt.subplots(1, 5, figsize=(15, 5))

for ax, (N, (inside, outside)) in zip(axes, zip(N_values, results)):
    pi_estimate = len(inside) / (len(inside) + len(outside)) * 4  # π 값 계산
    ax.scatter([x[0] for x in inside], [x[1] for x in inside], s=5, c='red', alpha=0.7)
    ax.scatter([x[0] for x in outside], [x[1] for x in outside], s=5, c='blue', alpha=0.7)
    ax.set_title(f'N = {N}\nπ ≈ {pi_estimate:.5f}')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')

# 그래프 출력
plt.tight_layout()
plt.show()
```
